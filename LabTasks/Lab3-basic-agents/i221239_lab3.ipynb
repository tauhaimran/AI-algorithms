```python
from random import randint
import numpy as np
#Environment:
#A 3x3 grid where each cell contains plants (Dry/Wet).
class plant_grid:
  def __init__(self,grid,Wets,Drys):
    self.grid = grid
    self.m = len(grid)
    self.n = len(grid[0])
    self.Wets = self.count_wet()
    self.Drys = self.count_dry()

  def count_wet(self):
    count = 0
    for i in range(self.m):
      for j in range(self.n):
        if self.grid[i][j] == 'Wet':
          count += 1
    return count

  def count_dry(self):
    count = 0
    for i in range(self.m):
      for j in range(self.n):
        if self.grid[i][j] == 'Dry':
          count += 1
    return count

  def print_grid(self):
    for i in range(self.m):
      for j in range(self.n):
        print(self.grid[i][j], end=' ')
      print()
    print ('Dry->',self.Drys,' Wet->',self.Wets)

  def get_percept(self,x,y):
    return self.grid[x][y]

  def update_grid(self,x,y,new_value):
    old_value = self.grid[x][y]
    self.grid[x][y] = new_value
    if new_value == 'Wet' and old_value == 'Dry':
      self.Wets += 1
      self.Drys -= 1
    elif new_value == 'Dry' and old_value == 'Wet':
      self.Drys += 1
      self.Wets -= 1

  def fully_Wet(self):
    if self.Wets == 9 and self.Drys == 0:
      return True
    else:
      return False

#Goal Based Agent
class GoalBasedAgent:
  def __init__(self, xpos , ypos,dirnum, direction):
      self.xpos = xpos
      self.ypos = ypos
      self.dirnum = dirnum
      self.direction = direction


  def chose_direction(self):
    self.dirnum = (self.dirnum + 1) % 4 # Use modulo to cycle through directions
    if self.dirnum == 0:
      self.direction = 'Right'  # Start with 'Right'
    elif self.dirnum == 1:
      self.direction = 'Down'
    elif self.dirnum == 2:
      self.direction = 'Left'
    elif self.dirnum == 3:
      self.direction = 'Up'

  def move(self,grid,x,y):

    # Store previous position
    prev_x, prev_y = self.xpos, self.ypos

    if self.direction == 'Left':
      self.xpos = x - 1
    elif self.direction == 'Right':
      self.xpos = x + 1
    elif self.direction == 'Up':
      self.ypos = y + 1
    elif self.direction == 'Down':
      self.ypos = y - 1

    # Check if move is valid, revert if not
    if self.xpos < 0 or self.xpos >= grid.m or self.ypos < 0 or self.ypos >= grid.n:
      self.xpos, self.ypos = prev_x, prev_y  # Revert to previous position
      self.chose_direction()                # Choose a new directions >= grid.m or self.ypos < 0 or self.ypos >= grid.n:
      #self.chose_direction()

  def print_direction(self):
    print("> moving ... ",self.direction)

  def formulate_goal(self,plant_grid):
    while(plant_grid.fully_Wet() == False):
      self.print_direction()
      self.move(plant_grid,self.xpos,self.ypos)
      percept = plant_grid.get_percept(self.xpos,self.ypos)
      new_state = self.decide(percept)
      plant_grid.update_grid(self.xpos,self.ypos,new_state) #Use new_state here
      print ('Current Grid')
      plant_grid.print_grid()
    print('>> Goal Reached')

  def decide(self, percept):
    if percept == 'Wet':
      print('>> Already Wet Moving Ahead... ')
    elif percept == 'Dry':
      print('>> Watering Plant... ')

    return 'Wet'


# Driver code
A = np.array([['Dry','Dry','Dry'],
              ['Dry','Wet','Dry'],
              ['Dry','Dry','Dry']])
grid = plant_grid(A,0,0)
print ('Current Grid')
grid.print_grid()
agent = GoalBasedAgent(0,0,0,'Right') # instantiate GoalBasedAgent
agent.formulate_goal(grid)
```
