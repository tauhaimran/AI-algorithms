```python
# series
import pandas as pd
data=[1,2,3,4,5]
data=pd.Series(data)
print(data)
```

```python
# selecting multiple columns
df[["date","price"]]
```

```python
# filter the data based on come condition
high_price = df[df["price"] > 1000000]
print(high_price.shape)
```

```python
#sorting the dataframe based on a price in decending order
high_price.sort_values("price",ascending=False) 
```

```python
high_price
```

```python
# selecting the row based on the actual position
high_price.iloc[1]
```

```python
# making new features and adding a column 
df["tax"]=df["price"]*0.10
df.head()
```

```python
# drop a column 
df.drop("taxed_price",axis=1,inplace=True)
df.head()
```

```python
# droping rows where there is null values
df.dropna(inplace=True)

# replace null values with other value
df.fillna(1,inplace=True)

# droping row based on the null value in a specified column
df.dropna(subset=['Date'], inplace = True)
```

```python
# make average of column in a dataframe
df.groupby('floors').mean(numeric_only=True)
```

```python
# reset the index drop false remove the old indexes without keeping them 
df.reset_index(drop=False, inplace=False)
```

```python
# detecting the duplicates
df.duplicated()
```

```python
#remove the duplicates 
df.drop_duplicates()
```

```python
# print the rows which are duplicated 
df[df.duplicated()]
```

```python
# keep the last row of the duplicates
df = df.drop_duplicates(keep='last')
```

```python
df.dtypes
```

```python
# correlation in columns
columns_for_corr=[]
for column in df.columns:
    if column != "date":
        columns_for_corr.append(column)
print(columns_for_corr)


df[columns_for_corr].corr()
```

```python
# finding the correlation of columns with the target variable
df.drop("date",axis=1).corr()['price']
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
columns=df.columns
corr=df[columns[4:7]].corr()
plt.figure(figsize=(12, 8))
sns.heatmap(corr, annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

```python
df[columns[4:7]].corr()
```

```python
# scatter plot in matplotliab
plt.figure(figsize=(10,5))
plt.scatter(x=df["bedrooms"],y=df["bathrooms"],color="red",marker="o",s=50,alpha=0.7,edgecolors="black")
plt.xlabel("bedrooms")
plt.ylabel("bathrooms")
plt.title("Bedrooms Vs Bathrooms Comparision")
plt.show()
```

```python
# line plot in matplotliab

plt.figure(figsize=(10,5))
plt.plot(df["bedrooms"],df["bathrooms"],color="red",marker="o",alpha=0.7)
plt.xlabel("bedrooms")
plt.ylabel("bathrooms")
plt.title("Bedrooms Vs Bathrooms Comparision")
plt.show()
```
